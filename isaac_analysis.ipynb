```python
import os

all_csv = [csv for csv in os.scandir('data') if ".csv" in csv.name]
all_csv
```

```python
import pandas as pd
angel = pd.read_csv(all_csv[2])
angel
```

```python
all_items = pd.read_csv(all_csv[1])
all_items
```

```python
def joinAll(df1 , all_df):
    df = pd.merge(df1, all_df, on=['Name','ID','Description'], how='left')
    return df
```

```python
test = joinAll(angel, all_items)
test.to_csv('sus.csv', index = False)
```

```python
def getQuality(all_csv, all_items):
    
    for i,csv in enumerate(all_csv):
        if all_csv[i] == 'all.csv':
            continue
        csv_df = pd.read_csv(csv)
        update = joinAll(csv_df, all_items)
        update.replace(10.5, 0.5)
        update.to_csv(csv, index = False)
        

getQuality(all_csv, all_items)
```
